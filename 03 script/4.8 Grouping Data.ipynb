```python
#Create Max order Column

ords_prods_merged['max_order'] = ords_prods_merged.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merged.head(100)
```

```python
pd.options.display.max_rows = None
```

```python
#Assign Loyalty customers

ords_prods_merged.loc[ords_prods_merged['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
#Regular customers

ords_prods_merged.loc[(ords_prods_merged['max_order'] <= 40) & (ords_prods_merged['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
#New customer

ords_prods_merged.loc[ords_prods_merged['max_order'] < 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merged['loyalty_flag'].value_counts(dropna= = False)
```

```python
#Data check

ords_prods_merged[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

```python
#5a. Spending habits

ords_prods_merged.groupby('loyalty_flag').agg({'prices':['mean', 'median', 'min', 'max']})
```

```python
#6a. Create 'spending' column

ords_prods_merged['spending'] = ords_prods_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#Create Spending Flag - Low Spender

ords_prods_merged.loc[ords_prods_merged['spending'] < 10, 'spending_flag'] = 'Low spender'
```

```python
#High spender

ords_prods_merged.loc[ords_prods_merged['spending'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merged['spending_flag'].value_counts(dropna = False)
```

```python
ords_prods_merged[['user_id', 'spending_flag', 'spending']].tail()
```

```python
#7a. Determine frequent customers. Create median days_since column 'frequency'

ords_prods_merged['frequency'] = ords_prods_merged.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
#non-frequent customer

ords_prods_merged.loc[ords_prods_merged['frequency'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
#Regular

ords_prods_merged.loc[(ords_prods_merged['frequency'] > 10) & (ords_prods_merged['frequency'] <= 20), 'order_frequency_flag'] = 'Regular customer'
```

```python
#Frequent customer

ords_prods_merged.loc[ords_prods_merged['frequency'] < 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
#check results

ords_prods_merged['order_frequency_flag'].value_counts(dropna = False)
```
